```python
import json 
import pandas as pd

file=open('stocks','r')
stockDict=json.load(file)
stocks=pd.DataFrame(stockDict['pageProps']['stocks'])
stocks.columns=['ticker','name','industry','market_cap']
stocks
```

```python
stocks=stocks[stocks.market_cap>0].copy()
stocks
```

```python
middle=int(len(stocks)/2)
```

```python
midCap=stocks.sort_values('market_cap').iloc[middle-50:middle+50]
midCap
```

```python
import yfinance as yf

tickers=' '.join(midCap.ticker)
data=yf.download(tickers=tickers,period='10y',interval='1d',auto_adjust=True)
data
```

```python
close=data['Close'].copy()
close
```

```python
l=[{'ticker':ticker,'n':len(close[ticker].dropna())} for ticker in close.columns]
lengths=pd.DataFrame(l)
lengths
```

```python
lengths.n.hist()
```

```python
lengths.n.mean()*.2
```

```python
import numpy as np

x=pd.Series(np.ones(10))
x.rolling(2)
```

```python
def test(series,short,long):
    x=series.to_frame(name='n')
    x['short']=x.n.rolling(short).sum()
    x['long']=x.n.shift(short).rolling(long).sum()
    return x
```

```python
test(pd.Series(np.arange(1,15,1)),2,3)
```
