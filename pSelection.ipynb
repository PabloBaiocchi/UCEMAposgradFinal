```python
import pandas as pd

data=pd.read_csv('large_cap.csv')
data.drop('Date',axis='columns',inplace=True)
data
```

```python
import random

indeces=random.sample(range(len(data.columns)),25)
dset=data.iloc[:,indeces]
dset
```

```python
import numpy as np
import strategy as s 

longRange=np.linspace(50,100,5,dtype=int)
shortRange=[int(l/3) for l in longRange]
pRange=[.0000000001,.000000001,.00000001,.0000001,.000001,.00001,.0001,.001,.01]

returns=[dset[ticker].pct_change() for ticker in dset.columns]
signals=[s.getBuySellSignals(r,shortRange,longRange,pRange,pRange) for r in returns]
```

```python
totalSignals=[]
for s in signals:
    totalSignals=totalSignals+s
```

```python
def countBuySell(signal):
    buy=0
    sell=0
    for s in signal['signal']:
        if s=='buy': 
            buy=buy+1
        elif s=='sell':
            sell=sell+1

    return {
        'lower_p': signal['lower_p'],
        'lower_short': signal['lower_short'],
        'lower_long': signal['lower_long'],
        'upper_p': signal['upper_p'],
        'upper_short': signal['upper_short'],
        'upper_long': signal['upper_long'],
        'buy_n':buy,
        'sell_n':sell
    }

rows=[countBuySell(signal) for signal in totalSignals]
df=pd.DataFrame(rows)
df
```

```python
lengths=[len(s['signal']) for s in totalSignals]
meanSignalLength=sum(lengths)/len(lengths)
meanSignalLength
```

```python
sellDf=df.groupby('upper_p').sell_n.mean().to_frame()
sellDf['sell_perc']=sellDf.sell_n/meanSignalLength
sellDf
```

```python
buyDf=df.groupby('lower_p').buy_n.mean().to_frame()
buyDf['buy_perc']=buyDf.buy_n/meanSignalLength
buyDf
```
